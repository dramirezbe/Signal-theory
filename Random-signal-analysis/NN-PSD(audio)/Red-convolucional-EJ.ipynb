```python
from tensorflow import keras
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, AveragePooling2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.regularizers import l2
from tensorflow.keras.callbacks import ModelCheckpoint


import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()
```

```python
(x_train,x_valid) = x_train[5000:], x_train[:5000]
(y_train,y_valid) = y_train[5000:], y_train[:5000]
```

```python
"""ONE HOT ENCODING"""

num_classes = 10
train_labels = to_categorical(y_train, num_classes)
test_labels = to_categorical(y_test, num_classes)
```

```python
"""Crear Modelo"""

base_filtros = 32
w_regularizer = 1e-3

from keras.api._v2.keras import regularizers

model = Sequential()

model.add(Conv2D(base_filtros*3,(3,3), activation= 'relu', kernel_regularizer = l2(5e-4),padding='same', input_shape = (32,32,3)))

#Second Convolutional layer

model.add(Conv2D(base_filtros*3,(3,3), activation= 'relu', kernel_regularizer = l2(5e-4),padding='same'))

#Third Convolutional layer

model.add(Conv2D(base_filtros*3,(3,3), activation= 'relu',padding='same'))

#Pooling Layer

model.add(MaxPooling2D(pool_size=(2,2)))

#Convolutional layer number 4

model.add(Conv2D(base_filtros*3,(3,3), activation= 'relu', kernel_regularizer = l2(5e-4),padding='same'))

#Convolutional layer number 5

model.add(Conv2D(base_filtros*3,(3,3), activation= 'relu', kernel_regularizer = l2(5e-4),padding='same'))

#Pooling Layer

model.add(MaxPooling2D(pool_size=(2,2)))

#Convolutional layer number 6

model.add(Conv2D(base_filtros*3,(3,3), activation= 'relu', kernel_regularizer = l2(5e-4),padding='same'))

#Convolutional layer number 7

model.add(Conv2D(base_filtros*3,(3,3), activation= 'relu', kernel_regularizer = l2(5e-4),padding='same'))

#Pooling Layer

model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Flatten())

model.add(Dense(128, activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
```

```python
"""Compilador"""

custom = SGD(learning_rate=0.02)
model.compile(metrics=['accuracy'], loss='sparse_categorical_crossentropy', optimizer='rmsprop')
```

```python
#Callback
checkpoint = ModelCheckpoint("best_model.h5",monitor = 'val_accuracy', save_best_only=True, mode='max', verbose=1)
```

```python
history = model.fit(x_train, y_train, batch_size=192, epochs=50, validation_data=(x_valid, y_valid), verbose=2, shuffle=True, callbacks=[checkpoint])
```
