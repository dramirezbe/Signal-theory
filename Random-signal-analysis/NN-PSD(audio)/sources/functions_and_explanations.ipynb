# Functions

## Plotter

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_func(x,y, title):
    # helper function to plot activation functions
    plt.plot(x, y)
    plt.title(title)
    plt.xlabel('x')
    plt.ylabel('activation(x)')
    plt.grid(True)
    plt.show()

x = np.linspace(-10, 10, 100)
```

## Sigmoid
- Maps the input value between 0 and 1 (Normalization)

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

y = sigmoid(x)
plot_func(x,y,'Sigmoid')
```

## Rectified Linear Unit (ReLU)

- Does the negative values 0, and otherwise not

```python
def relu(x):
    return np.maximum(0, x)

y = relu(x)
plot_func(x,y,'ReLU')
```

## Leaky ReLU

- The Leaky ReLU activation function is a variation of the ReLU function that allows a small, non-zero output for negative input values, preventing complete suppression of information.

```python
def leaky_relu(x, alpha=0.1):
    return np.maximum(alpha*x, x)

y = leaky_relu(x)
plot_func(x,y,'Leaky ReLU')
```

## Tanh

```python
def tanh(x):
    return np.tanh(x)

y = tanh(x)
plot_func(x,y,'Tanh')
```

## Softmax

- Converts a real vector in a probabilty distribution (used in multiclass classification)

```python
def softmax(x):
    exp_scores = np.exp(x)
    return exp_scores / np.sum(exp_scores)


y = softmax(x)
plot_func(x,y,'Softmax')
```

# Explanations

## Multiple layers

- Using only ReLU would amplify the error. For this reason to adress the issue we can use a neural network with two layers.
- Two neural network formula:
$\hat{y} = w_2*relu(w_1*x+b_1)+b_2$

```python
prediction = lambda x, w1=.2, b=1.99: x*w1 + b


#Apply ReLU activation function
fig, axs = plt.subplots(1, 3, figsize = (15,3))

layer1_1 = relu(prediction(x))
axs[0].plot(x, layer1_1)
axs[0].set_title('layer1_1')

layer1_2 = relu(prediction(x, .3, -2))
axs[1].plot(x,layer1_1+layer1_2)
axs[1].set_title('layer1_1+layer1_2')

layer1_3 = relu(prediction(x, .6, -2))
axs[2].plot(x,layer1_1+layer1_2+layer1_3)
axs[2].set_title('layer1_1+layer1_2+layer1_3')

plt.show()
```
