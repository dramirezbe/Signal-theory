```python
import numpy as np
import matplotlib.pyplot as plt
from librosa import load, get_duration
from scipy.signal import welch, butter, sosfilt, sosfreqz
```

## PSD de Bohemian-Rapsody sin ecualizar

```python
audio, sr = load('sources/Bohemian_Rhapsody.wav', sr=None) #sr=sample_rate Hz

# Asegurar audio en mono
if audio.ndim > 1:
    audio = np.mean(audio, axis=1)

duration = get_duration(y=audio, sr=sr) #seconds

n = int(duration * sr)

t = np.arange(n) / sr
```

```python
frequencies, psd = welch(audio, sr, nperseg=1024)

fig, axs = plt.subplots(1, 2, figsize = (15,3))
axs[0].plot(t,audio)
axs[1].plot(frequencies,psd)
plt.show()
```
