### Ejercicio 1: Sistema de Biblioteca
Crea una clase Libro que contenga atributos como título, autor, año de publicación y estado (disponible o prestado). Luego, desarrolla una clase Biblioteca que gestione una colección de objetos Libro. Esta clase debe incluir métodos para:

Añadir un nuevo libro al catálogo.
Prestar un libro a un usuario (cambiar el estado a prestado).
Devolver un libro (cambiar el estado a disponible).
Buscar libros por autor o título.
Mostrar los libros disponibles para préstamo.

```python
class Book:
    def __init__(self, title, author, year, state = "Disponible"):
        self.title = title
        self.author = author
        self.year = year
        self.state = state

class Library:
    def __init__(self):
        self.catalog = []

    def add(self, book):
        self.catalog.append(book)

    def lend(self, book_title):
        for book in self.catalog:
            if book.title == book_title:
                if book.state == "Disponible":
                    book.state = "Prestado"
                    print(f"The book '{book.title}' has been lent.")
                else:
                    print(f"The book '{book.title}' is not available for lending.")
                return
        print(f"The book '{book_title}' is not in the library's catalog.")

    def return_book (self, book_title):
        for book in self.catalog:
            if book.title == book_title:
                if book.state == "Prestado":
                    book.state = "Disponible"
                    print(f"The book '{book.title}' has been returned.")
                else:
                    print(f"The book '{book.title}' is already in tle library.")
                return
        print(f"The book '{book_title}' is not in the library's catalog.")

    def search(self, book_title):
        for book in self.catalog:
            if book_title not in book:
                print("Your book doesn't exist in the catalog")
        






book1 = Book("Harry Potter 1", "JK Rowling", 1997)
book2 = Book("Hail Marry Project", "Andy Weir", 2021)
book3 = Book("Harry Potter 2","JK Rowling","1998")

library = Library()
library.add(book1)
library.add(book2)
library.add(book2)





```

### Ejercicio 2: Sistema de Reservas de Hotel
Define una clase Habitacion con atributos como número de habitación, tipo (simple, doble, suite), precio por noche y estado (ocupada o disponible). Luego, crea una clase Hotel que administre varias habitaciones. La clase Hotel debe ofrecer funcionalidades para:

Registrar una nueva habitación.
Reservar una habitación, cambiando su estado a ocupado.
Liberar una habitación después de la estancia, cambiando su estado a disponible.
Calcular el costo de la estancia de un cliente.
Mostrar todas las habitaciones disponibles.

```python
class Room:
    def __init__(self, number, type, price_per_night):
        self.number = number
        self.type = type
        self.price_per_night = price_per_night
        self.state = "available"

class Hotel:
    def __init__(self):
        self.rooms = []


    def register(self, new_room):
        self.rooms.append(new_room)
        print("OK")
        print("Habitaciones actualizadas: ",self.rooms)

    def reserve(self,reserve_room):
        for loop_room in self.rooms:
            if loop_room.number == reserve_room:
                if loop_room.state == "available":
                    loop_room.state = "occupied"
                    print("Proceso exitoso")
                else:
                    print("Habitación ocupada, intente con otra")
    
    def vacate(self, vacate_room):
        for loop_room in self.rooms:
            if loop_room.number == vacate_room:
                if loop_room.state == "occupied":
                    loop_room.state = "available"
                    print("Proceso exitoso")
                else:
                    print("Habitación ya disponible, intente con otra")

    def cost_client(self, nroom, days):
        for loop_room in self.rooms:
            if loop_room.number == nroom:
                cost = loop_room.price_per_night * days
                print("Deuda a pagar: ", cost)
                return

                 
        

hotel = Hotel()

room1 = Room(1, "simple", 100)
room2 = Room(2, "double", 200)
room3 = Room(3, "suite", 300)

hotel.register(room1)
hotel.register(room2)
hotel.register(room3)
```
